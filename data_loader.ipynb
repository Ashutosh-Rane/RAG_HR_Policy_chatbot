```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
documents = PyMuPDFLoader('data\HR_Policies.pdf')
```

```python
documents.load()
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(documents.load())
print(f'Total chunks: {len(chunks)}')
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="nomic-embed-text")
```

```python
from langchain_community.vectorstores import FAISS
```

```python
vestorstore = FAISS.from_documents(chunks, embeddings)
```

```python
results = vestorstore.similarity_search(
    "What is maternity leave policy?",
    k=3
)

for r in results:
    print(r.page_content[:500])
```

```python
vestorstore.save_local("faiss_index")
```
